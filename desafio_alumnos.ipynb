Importa el archivo alumnos.csv y almacena su contenido en un DataFrame de Pandas.

Extra: Calcula algunas estadísticas descriptivas básicas de los datos en el DataFrame (media, desviación estándar, etc.). Pista: busca el método "describe".

```python
# Calcular estadisticas descriptivas de valores int64 y float64

estadisticas_descriptivas = datos.describe()
estadisticas_descriptivas
```

```python
total_alumnos = datos['Nombre'].count()
print(f"La cantidad de alumnos es: {total_alumnos}")

alumnos_sin_nota = datos['Nota'].isnull().sum()
print(f"La cantidad de alumnos sin nota es: {alumnos_sin_nota}")

alumnos_con_nota = datos['Nota'].count()
print(f"La cantidad de alumnos con nota es: {alumnos_con_nota}")

edad_media = datos['Edad'].mean()
print(f"La edad promedio es: {edad_media}")

edad_mas_baja = datos['Edad'].min()
print(f"La edad mas baja es: {edad_mas_baja}")

edad_mas_alta = datos['Edad'].max()
print(f"La edad mas alta es: {edad_mas_alta}")
```
